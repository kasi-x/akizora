```python
import re

def mark_sections(text):
    lines = text.split('\n')  # テキストを行ごとに分割
    modified_lines = []  # 変更後の行を格納するリスト

    for line in lines:
        line = line.strip()  # 行の冒頭と文末のスペースを削除
        # CHAPTERにマッチする場合
        if re.match(r'(?i)^CHAPTER\s+\d+', line):
            modified_lines.append(f'<CHAPTER><CHAPTER_TITLE> {line} </CHAPTER_TITLE>')
        # PARTにマッチする場合
        elif re.match(r'(?i)^PART\s+\d+', line):
            modified_lines.append(f'<PART><PART_TITLE> {line} </PART_TITLE>')
        # EPILOGUEにマッチする場合（大文字小文字を問わず）
        elif re.match(r'(?i)^EPILOGUE', line):
            modified_lines.append(f'<EPILOGUE> {line}')
        # THE ENDにマッチする場合（大文字のみ）
        elif re.match(r'^THE END$', line):
            modified_lines.append(f'<THE_END> {line} </THE_END>')
        # Openingにマッチする場合（大文字小文字を問わず）
        elif re.match(r'(?i)^OPENING', line):
            modified_lines.append(f'<OPENING> {line}')
        elif re.match(r'(?i)^PROLOGUE', line):
            modified_lines.append(f'<PROLOGUE><PROLOGUE_TITLE> {line} </PROLOGUE_TITLE>')
        elif re.match(r'(?i)^TRANSCRIBER NOTES', line):
            modified_lines.append(f'<TRANSCRIBER_NOTES> {line} ')
        # その他の行
        else:
            modified_lines.append(line)

    # 変更後のテキストを改行で結合して返す
    return '\n'.join(modified_lines)
```

```python
import re

def remove_single_newlines(text):
    # 各行を処理するためにテキストを分割
    lines = text.split('\n')
    modified_lines = []  # 変更後の行を保持するリスト

    for line in lines:
        # タグを含む行はそのまま追加
        if '<' in line and '>' in line:
            modified_lines.append(line)
        else:
            # 単独の改行の前後に文字がある場合のみ、行末の改行を空白に置換
            line = re.sub(r'(?<!\n)\n(?!\n)', ' ', line)
            modified_lines.append(line)
    
    # 連続する改行はそのまま保持されるように、変更後の行を改行で結合
    return '\n'.join(modified_lines)
```

```python
def auto_close_tags(text):
    priority = {
        "PART": 3,
        "CHAPTER": 2,
        "TRANSCRIBER_NOTES": 4,
        "THE_END": 4,
        "EPILOGUE": 4,
        "PROLOGUE": 4,
        "DATA": 1,
    }
    priority = dict(sorted(priority.items(), key=lambda item: item[1]))

    def sort_tags(tags):
        return sorted(tags, key=lambda tag: priority.get(tag, -1))

    def get_priority(tag):
        return priority.get(tag, -1)

    open_tags = []
    output = []
    remove_list = []

    lines = text.split("\n")
    for line in lines:
        for new_tag in priority:
            start_tag = f"<{new_tag}>"
            end_tag = f"</{new_tag}>"
            if start_tag in line:
                if open_tags:
                    for open_tag in open_tags:
                        if get_priority(open_tag) <= get_priority(new_tag):
                            output.append(f"</{open_tag}>")
                            remove_list.append(open_tag)
                    open_tags = [new_tag for new_tag in open_tags if new_tag not in remove_list]
                    remove_list = []
                open_tags.append(new_tag)
                open_tags = sort_tags(open_tags)

            if end_tag in line:
                open_tags.remove(new_tag)
                if open_tags:
                    for open_tag in open_tags:
                        if get_priority(open_tag) <= get_priority(new_tag):
                            output.append(f"</{open_tag}>")
                            remove_list.append(open_tag)
                    open_tags = [new_tag for new_tag in open_tags if new_tag not in remove_list]
                    remove_list = []
        output.append(line)

    while open_tags:
        output.append(f"</{open_tags.pop()}>")

    return "\n".join(output)
```

```python
def close_open_sections_and_chapters(processed_lines, section_open, chapter_open):
    if section_open:
        processed_lines.append('</SECTION>')
    if chapter_open:
        processed_lines.append('</CHAPTER>')
    return processed_lines, False, False

def start_new_chapter(processed_lines, section_open, chapter_open):
    processed_lines, section_open, chapter_open = close_open_sections_and_chapters(processed_lines, section_open, chapter_open)
    chapter_open = True
    processed_lines.append('<CHAPTER>')
    return processed_lines, section_open, chapter_open

def start_new_section(processed_lines, section_open):
    if section_open:
        processed_lines.append('</SECTION>')
    section_open = True
    processed_lines.append('<SECTION>')
    return processed_lines, section_open

def process_text(text):
    lines = text.split('\n')
    processed_lines = []
    chapter_open = False
    section_open = False

    for line in lines:
        if '<CHAPTER_START>' in line:
            processed_lines, section_open, chapter_open = start_new_chapter(processed_lines, section_open, chapter_open)
        elif '<SECTION_POINT>' in line:
            processed_lines, section_open = start_new_section(processed_lines, section_open)
        else:
            processed_lines.append(line)

    # 文書の最後で開いているセクションと章を閉じる
    processed_lines, section_open, chapter_open = close_open_sections_and_chapters(processed_lines, section_open, chapter_open)

    return '\n'.join(processed_lines)
```
